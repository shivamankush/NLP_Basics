# A Complete Guide to NLP: From Text to Insights

## Introduction

Natural Language Processing (NLP) is a subfield of Computer Science,Linguistics,Artificial Intelligence concerned with the interactions between human language and computers,in particular how to program computers to process and analyze large amounts of natural language data..  Example->Chatbots,Siri,etc..  

## Common NLP Tasks

1. **Text/Document Classification**: Assume a text of news and our task is to assign category to it(i.e.,it is of sports or politics or entertainment). This task can be easily done using text classifiacation..

2. **Sentiment Analysis**: We take reviews of public or consumers on social media or e-commerce site about movie or particular product respectively and by extracting the emotions from the reviews We will be able to know that what they are feelig about the movie or product..

3. **Information Retrieval**: Search Engines internally use information retrieval.In this we extract entities such as date,product name,event name,etc from text..

4. **Parts of Speech Tagging**: It is very important text preprocessing step where we assign 'parts of speech'(whether it is Noun,Verb,Adjective,etc) to every word of text ..

5. **Language Detection and Machine Translation**: Firstly we detect in which language user is giving input and based on in which language user prefer the output,being tranlated to that language.It is used in "Google translate"..

6. **Conversational Agents**: Conversational agents, also known as "chatbots" or "virtual agents", are AI-powered programs designed to engage in natural language conversations with users.. It can be of two types -- 1.Text Based(Chatbot of Swiggy)|| 2. Speech Based(Siri)..

7. **Knowledge Graph and Q&A Systems**: If we have a very large dataset then we can build "Knowledge Graph" by logically connecting the entities with one another. Finally we can convert the Knowledge Graph into Q&A system.. In reality Google actually use this technology.. 

8. **Text Summarization**: We can summarize a full fleged article to give it a short form. InShorts use this technique to give news summary in less than 60 words related to business,sports,bollywood and technology in india and around the world in both english and hindi..

9. **Text Generation**: As we are typing something then on basis of previous typing behaviour it predicts that what should be the next word..

10. **Text Parsing**: Parsing, also known as "syntactic analysis", is the process of analyzing a sequence of tokens to determine the grammatical structure of a program..
                  Example:  Sentence: "The cat sat on the mat."
                            Parsing: The parser would identify "The cat" as a noun phrase, "sat" as a verb, and "on the mat" as a prepositional phrase. 

## Approaches to NLP

1. **Heuristic Method**(1950s-1990s):
 
a.**Regular Expressions**:Regular expressions, often shortened to "regex," are patterns used to match character combinations in strings, enabling powerful text searching, validation, and manipulation tasks.Patterns: Regex are sequences of characters that define a search pattern. 
Used for:
Searching: Finding specific patterns within text. 
Validation: Ensuring data conforms to a predefined format (e.g., email, phone number). 
Extraction: Isolating specific parts of text. 
Replacement: Modifying text based on patterns. 
Ubiquitous: Used in many programming languages and tools. 

b. **WordNet**(Lexical Dictionary): In this We store the possible relationship of every word with other words in an organised manner so that in future if we want to build a complex NLP system then Wordnet or Lexical dictionary will help us in doing so..

c.**Open Mind Common Sense**: Open Mind Common Sense (OMCS) is a crowdsourced knowledge base project aiming to provide AI systems with common-sense knowledge, crucial for NLP tasks like understanding context and generating realistic narratives, by leveraging human understanding of the world. 
  #OMCS is a project at MIT Media Lab that aims to build a large, crowd-sourced common sense knowledge base. 
  #It allows people to contribute natural language statements about the world, forming a database of common-sense knowledge. 
  #The goal is to give AI systems access to this knowledge, enabling them to understand and reason about the world like humans. 


2. **Machine Learning Based Methods**(1990s-till now):
   
  In Heuristic Approaches We make general rules but in reality there are some open ended problems where we can't make general rules so there comes ML algorithm into picture that make rules internally by seenig the relationship between input and output of data..
ML Workflow for solving NLP problems: We convert textual data into numbers(i.e.,called Text Vectorization) and then we sent that numbers into ML algorithm to give some output and then we evaluate the result..
ML Algo. used:
**Naive Bayes**--Gives very good result on textual data

--The problem in ML based approaches is that we have to convert the textual data into numbers and in doing so most of the time we loose the sequential information.It is because ML Algo are desigened in sauch a way that they don't care about the sequential information BUT Deep Learning methods use this sequential information(like sentence formation takes place from left to right) and this is huge advantage of DL based models that it is able to store the "Sequential information"..

**Convolutional Neural Network(CNN)**--Mainly used for image classification..
**Gated Recurrent Unit(GRU)**--It is mostly used for text generation..
**Transformers**--Provides attention on specific parts of the sentence..

## Challenges in NLP:
1. **Ambiguity in Language**
Lexical Ambiguity: Words with multiple meanings (e.g., bat can mean an animal or sports equipment).

Syntactic Ambiguity: Different sentence structures (e.g., He saw the man with the telescope—who has the telescope?).

Semantic Ambiguity: Different interpretations of meaning (e.g., I saw a man on a hill with a telescope).

2. **Context Understanding**
Words and sentences depend on the context, making understanding difficult.

Example: "That’s cool!" could mean temperature or excitement.

3. **Sarcasm and Irony**
Sentiment analysis struggles with sarcasm (e.g., "Oh great, another Monday!" sounds positive but is usually negative).

4. **Named Entity Recognition (NER) Issues**
Identifying names, places, and organizations can be difficult, especially for new or uncommon words (e.g., recognizing "Tesla" as a company vs. a scientist).

5. **Language Variability & Slang**
Dialects, slang, and informal speech make NLP models struggle (e.g., "gonna" vs. "going to").

6. **Low-Resource Languages**
Many NLP advancements are focused on English, while low-resource languages lack sufficient training data.

7. **Handling Misspellings & Typing Errors**
NLP models need to correct and interpret typos (e.g., "recieve" → "receive").

8. **Ethical Issues & Bias**
NLP models can inherit biases from training data (e.g., gender, racial, or political biases in AI responses).

9. **Long-Range Dependencies**
Some sentences require understanding distant words (e.g., "The book that I bought last week was amazing.").

10. **Real-Time Processing**
Implementing NLP in real-time applications (like chatbots or speech recognition) requires high speed and accuracy.

## Removing Punctuations

When we are tokenizing the sentences, punctuations become different words, ‘for example Hello! how are you?’ when we tokenize it can become Hello , !, how ,are, you,? here punctuation marks are taken as different words. It adds complexity to the model by performing more operations to the code and punctuations does not attribute any meaning to the sentence..

```python
import string
string.punctuation
```

```python
exclude=string.punctuation
```

```python
def remove_punc(text):
    for char in exclude:
        text=text.replace(char,"")
    return text
```

## Stop words Removal

Words which help in sentence formation, but does not add any meaning to the sentence like ‘a’, ‘an’, ‘the’, ‘is’ and so on..

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
def remove_stop_words(text):
    new_text=[]
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    return ' '.join(new_text)
```

## Handling emoji’s

While analyzing chatting messages, for text classification or sentiment analysis, emoji’s particularly have a meaning, we can replace emoji expression with their corresponding meaning. If our problem statement does not have anything to do with sentiments or feelings, we can remove them..


**Removing emoji’s:**

```python
import re
def emoji_removal(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
emoji_removal('loved the movie it was ❤️❤️')
```

## Stemming

**Stemming is the process of reducing inflection in words to their root forms such as mapping a group of words to the same stem even if the stem itself is not a valid word in the Language..**

Inflection is the modification of a word to express different grammatical categories such as tense, voice, aspect, person, gender and mood..

In stemming, inflection is not cared much, other than the reducing the word to its stem..

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

## Lemmatization

**Lemmatization unlike stemming reduces inflected words properly ensuring that the root word belongs to the language. In the lemmatization root word is called lemma..**

```python
from nltk.stem import WordNetLemmatizer
wnl=WordNetLemmatizer()
def lemmatizing(text):
    return " ".join([wnl.lemmatize(word,pos='v') for word in text.split()])
lemmatizing("walks walking walked walk probably my all time favori swimming and walking")
```

To build the corpus from the 'verified_reviews' we perform the following -

1.Replace any non alphabet characters with a space

2.Covert to lower case and split into words

3.Iterate over the individual words and if it is not a stopword then add the stemmed form of the word to the corpus

```python
corpus = []
stemmer = PorterStemmer()
for i in range(0, data.shape[0]):
  review = re.sub('[^a-zA-Z]', ' ', data.iloc[i]['verified_reviews'])
  review = review.lower().split()
  review = [stemmer.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```
